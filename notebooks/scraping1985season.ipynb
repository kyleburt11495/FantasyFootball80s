```python
def getTeamUrls(url):
    page = requests.get(url)
    contents = page.content
    soup = BeautifulSoup(contents, 'html.parser')
    table_headings = soup.find_all('th', {'data-stat':'team'})
    team_hrefs = []
    for heading in table_headings:
        if heading.find_all('a') != []:
            for a in heading.find_all('a'):
                team_hrefs.append(a.get('href'))
    return team_hrefs

def getPlayerUrls(team_url):
    page = requests.get("https://www.pro-football-reference.com" + team_url.split(".htm")[0] + "_roster.htm")
    contents = page.content
    soup = BeautifulSoup(contents, 'html.parser')
    td_tags = soup.find_all("td", {'data-stat':'player'})
    print(td_tags)
    player_urls = set()
    for td in td_tags:
        if td.find('a') != None:
            player_urls.add(td.find('a').get('href'))
    
    
    

```

```python
team_urls = getTeamUrls('https://www.pro-football-reference.com/years/1985/') 
team_urls
for url in team_urls:
    getPlayerUrls(url)
```
