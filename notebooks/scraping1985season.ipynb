```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver

path_to_chromedriver = "/home/kyle/Selenium/chromedriver"
browser = webdriver.Chrome(executable_path = path_to_chromedriver)
```

```python
def getTeamUrls(url):
    page = requests.get(url)
    contents = page.content
    soup = BeautifulSoup(contents, 'html.parser')
    table_headings = soup.find_all('th', {'data-stat':'team'})
    team_hrefs = []
    for heading in table_headings:
        if heading.find_all('a') != []:
            for a in heading.find_all('a'):
                team_hrefs.append(a.get('href'))
    return team_hrefs

def getTeamRoster(team_url):
    team_roster = set()
    browser.get("https://www.pro-football-reference.com" + team_url.split(".htm")[0] + "_roster.htm")
    td_tags = browser.find_elements_by_css_selector("td[data-stat='player']")
    for td in td_tags:
        a_tags = td.find_elements_by_tag_name('a')
        for a in a_tags:
            team_roster.add(a.get_attribute("href"))
    return team_roster

def getPlayerInfo(player_url):
    print(player_url)
    playerDict = {}
    page = requests.get(player_url)
    contents = page.content
    soup = BeautifulSoup(contents, 'html.parser')
    name = soup.find("h1", {'itemprop':'name'})
    playerDict['name'] = name.string
    playerDict['id'] = player_url
    #find players position
    p_tags = soup.find_all('p')
    for p in p_tags:
        strong_tags = p.find_all("strong")
        if strong_tags != []:
            if strong_tags[0].string == 'Position':
                position = p.get_text()
                playerDict['position'] = getPlayerPosition(position)
                return playerDict
    return playerDict

def getPlayerPosition(position):
    #reads in string of player position and returns standard string
    if "RB" in position or "FB" in position:
        return "RB"
    elif "WR" in position:
        return "WR"
    elif "QB" in position:
        return "QB"
    elif "TE" in position:
        return "TE"
    elif "K" in position:
        return "K"
    else:
        return "NON-FANTASY"
```

```python
team_urls = getTeamUrls('https://www.pro-football-reference.com/years/1985/')
roster_list = []
for url in team_urls:
    print(url)
    roster_list.append( (url,getTeamRoster(url)) )
```

```python
new_roster_list = []
for team, roster in roster_list:
    roster = list(roster)
    new_roster_list.append( (team,roster) )
    
    
```

```python
roster_list = new_roster_list
```

```python
player_list = []
for team, roster in roster_list:
    for player in roster:
        player_dict = getPlayerInfo(player)
        player_dict['team'] = team
        if player_dict['position'] != "NON-FANTASY":
            player_list.append(player_dict)
```
